```python
import pandas as pd

# 1) ÌÉÄÏûÖ Î≥¥Ï†ï
ads_info["click_date"] = pd.to_datetime(ads_info["click_date"], errors="coerce")

# 2) ÎÇ†ÏßúÎßå Î∂ÑÎ¶¨ (Ï¥à/Î∂Ñ Ï†úÍ±∞)
ads_info["date"] = ads_info["click_date"].dt.date

# 3) click_timeÏù¥ 0~23Ïù∏ÏßÄ Ï†êÍ≤Ä(ÏÑ†ÌÉù)
ads_info = ads_info[ads_info["click_time"].between(0, 23)]

# 4) ÎÇ†Ïßú √ó ÏãúÍ∞ÑÎåÄÎ≥Ñ ÌÅ¥Î¶≠ Ïàò ÏßëÍ≥Ñ
daily_hourly = (
    ads_info.groupby(["date", "click_time"])
            .size()
            .reset_index(name="clicks")
)

# 5) ÎÇ†ÏßúÎ≥ÑÎ°ú 'ÌÅ¥Î¶≠ Ïàò ÏµúÎåÄ' ÏãúÍ∞ÑÎåÄÎßå Ï∂îÏ∂ú
idx = daily_hourly.groupby("date")["clicks"].idxmax()
peak_hours = daily_hourly.loc[idx].sort_values("date")

print("üìå ÎÇ†ÏßúÎ≥Ñ ÏµúÎã§ ÌÅ¥Î¶≠ ÏãúÍ∞ÑÎåÄ (Ï†ïÏÉÅ Í≥ÑÏÇ∞)")
print(peak_hours.head(10))
```

```python
overall_top10 = daily_hourly.sort_values("clicks", ascending=False).head(10)
print(overall_top10)
```

```python
# ÏãúÍ∞ÑÎåÄÎ≥Ñ(0~23) Ï†ÑÏ≤¥ ÌÅ¥Î¶≠ Ïàò Ìï©Í≥Ñ
hourly_clicks = daily_hourly.groupby("click_time")["clicks"].sum().reset_index()

# ÌÅ¥Î¶≠ Ïàò Í∏∞Ï§Ä ÎÇ¥Î¶ºÏ∞®Ïàú Ï†ïÎ†¨ ÌõÑ TOP 10
hourly_top10 = hourly_clicks.sort_values("clicks", ascending=False).head(10)

print("üìå Ï†ÑÏ≤¥ ÏãúÍ∞ÑÎåÄÎ≥Ñ ÌÅ¥Î¶≠ Ïàò TOP 10")
print(hourly_top10)
```
